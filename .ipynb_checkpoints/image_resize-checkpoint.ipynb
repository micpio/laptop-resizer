```python
#!bin/python3
import glob
from PIL import Image
import os, sys
f = os.getcwd()
fin = os.getcwd()+"/tobe/"
fout = os.getcwd()+"/processed/"
print(fin)
print(f)
def ProcessOne(f):
    files = (os.listdir(fin))
    # Create a list of files to process
     #files = [f for f in glob.glob("*.jpeg")]
    print("......................................................................")
    new_width = int(input("enter width desired"))
    print("......................................................................")
    for file in files:
            print("lets resize some images")
            print("Opening"+ file)
            im = Image.open(fin+file)
            if im.size >= (50,50): #process files that meet ext creds
                                     #minimum size to begin 300x300
                aspect_ratio = round((im.height / im.width),2)
                print("......................................................................")
                print("aspect ratio")
                print(aspect_ratio)
                print("......................................................................")
                new_height = round(new_width * aspect_ratio)
                new = im.resize((new_width, new_height))
                rf  = fout + "re" +file +".jpeg"  #format for file sent to processed
                if "re" not in f:
                    print("......................................................................")
                    new.save( rf)    #save
                    print(f)
                    print("originnal size")
                    print(im.size ) 
                    print("......................................................................")
                    print("resized with new name" )
                    print(rf)
                    print("new size")
                    print(new.size)
                    print("......................................................................")
                    print("Thank You ")
                    print("......................................................................")                
                else:
                    print("Thank You no more files available")
                    print("......................................................................")
                    
#if __name__ == '__main__':
                


ProcessOne(f)
```
